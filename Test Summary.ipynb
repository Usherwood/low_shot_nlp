```
import pandas as pd
import numpy as np
import csv
from sklearn.metrics import f1_score, accuracy_score
```

```
tasks = ['amazon_movies', 'amazon_health', 'amazon_books', 'semeval_task_a', 'semeval_task_ce']
```

```
accs = []
for task in tasks:
  df_tst = pd.read_csv("/content/gdrive/My Drive/Data Science/Low Shot NLP/datasets/"+task+"/fair_balanced_100/test.csv")
  current_accs = []
  for col in df_tst.iloc[:,2:].columns:
    current_accs.append(accuracy_score(df_tst['label'], df_tst[col]))
  accs.append(current_accs)
  
df = pd.DataFrame(np.array(accs).T, columns=tasks, index=df_tst.iloc[:,2:].columns)
```
